__1) Transformation of variables:__
For a random uniform distribution of x between 0 and 1, the probability density function is f(x). Find the probability density function g(u) of transformation: $$u=e^{−x}$$

a. Show details to derive the formula g(u).

$$g(u) = f(x)|\frac{dx}{du}|$$
$$u=e^{−x}$$
$$x = -\ln{u}$$
$$\frac{dx}{du} = -\frac{1}{u}$$ and distribution is uniform, so: $$f(x) = 1 $$
$$g(u) = 1*|-\frac{1}{u}| = \frac{1}{u} = g(u)$$


b. Draw three figures: transformation function u(x), f(x) and g(u)

```python
# importing necessary packages to use later
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

# defining an array of points between the specified endpoints to put into each function
# f(x) is just uniform distribution from 0 -> 1
f_ = (uniform.rvs(size=1000))

# points for the u(x) plot, by plugging into u(x), and linspace for plotting later
u_ = np.exp(-f_)
u_linspace = np.linspace(0.1,1.1,1000)

# plotting all 3 functions
# plotting the normalized random uniform distributed data
plt.hist(f_, bins = 100, alpha = .5, density = True, color = 'dodgerblue')
plt.xlabel('Randomly Generated Values')
plt.ylabel('Probability')
plt.title('Normalized Histogram for $f(x)$ of 1K points')
plt.show()

# plotting the transformed distribution data along with g(u)
plt.hist(u_, bins = 100, alpha = .5, range = [.3,1], density = True, color = 'lightskyblue')
plt.plot(u_linspace,(1/u_linspace), color = 'navy', lw = 2)

# adding details and labels, and legend
plt.ylim(0,7)
plt.xlabel('Randomly Generated Values')
plt.ylabel('Probability')
plt.title('Normalized Histogram for $u(x)$ of  the 1K points')
plt.xlim(.35,1.01)
plt.legend(['$g(u)$','$u(x)$'])
```

In order to get f(x), we randomly select a 1000 sample se from the uniform distribution and that is what the top plot is showing. It is the pdf of the 1000 points sampled, and that set of data is put into u(x) in order to transform it. The second plot shows what the transformed data looks like (light blue) while the dark blue curve fit shows the g(u) to fit the data that was transformed. 

__2) Rolling two 6-sided dices.__

a. What is the probability of rolling snake eyes (two 1s)?

We know that each die has an equal probability of landing on any of the faces, where there are 6 faces, so the probability of landing on a face is 1/6. To calculate the porobability of rolling two dice that are both snakes eyes, we multiply the probabilities of each die being a 1 to get the total probability. 
Die 1 probability of a 1 is $P = 1/6$ and Die 2 probability of a 1 is $P = 1/6$. This follows the probability equation $P = (\frac{1}{6})^{n}$ for the probability of rolling a certain sequence of dice. Thus, $P(1,1) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$. Thus, the probability of rolling snake eyes is $\frac{1}{36} = 0.028$.

b. What is the probability of rolling fives (dice sum to five)?



When considering the probability of rolling dice sum to five, we must consider the possible combinations of dice faces that will sum to five. This will be (1,4), (2,3) where either of the dice could be the 1 or 4, and the 2 or 3, so we should consider it as:
$$P(1,4) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
$$P(4,1) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
$$P(2,3) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
$$P(3,2) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
$$P(1,4) + P(4,1) + P(2,3) + P(3,2) = \frac{4}{36} = \frac{1}{9}$$
Thus, the probability of rolling fives is $\frac{1}{9} = 0.111$.

c. Rolling dices 5 times and get at least one snake eye. Is the probability larger than rolling fives?



We know that there is only one scenario of hitting snake eyes, which is $P(1,1) = \frac{1}{36}$ thus the probability of not hitting snake eyes on a roll is $\frac{35}{36}$. Therefore, if we roll twice and don't hit snake eyes, the probability is $\frac{35}{36}*\frac{35}{36} = (\frac{35}{36})^{2}$. This repeats for however many rolls you have, so the probability of not hitting snake eyes is $(\frac{35}{36})^{n}$. Therefore, in the case where we require hitting snake eyes at least once, that would just be the complement for not rolling snake eyes, which can be taken as $1 - (\frac{35}{36})^{n}$ for n rolls. For this scenario, $1 - (\frac{35}{36})^{n}$ where $n = 5$ so we have $1 - (\frac{35}{36})^{5}$.
$$1 - 0.8686157861 = .131$$
Therefore, the probability of rolling snake eyes at least once on 5 dice rolls is 0.131. The probability of rolling snake eyes at least once on 5 dice rolls is larger than the probability of rolling 2 numbers that sum to 5 on a single dice roll (0.131 > 0.111).

__3) Compare probability distribution functions__

Three probability distribution functions: Binomial, Poisson and Normal function.

__a. Binomial (N=10, p=0.02), Poisson (λ=0.2), Normal(μ=0.2, σ=0.45)__

i) Draw 3 functions in the same plot, and scale the area under the curve of each fucntion to 1. 

ii) Create a table to compare numerical values of Mean, Sigma, Skewness and Kurtosis, e.g. 

```python
# necessary imports to use later
import numpy as np
import scipy as sp
import pandas as pd
import random
import statistics as st
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import kurtosis
from scipy.stats import skew

# binomial distribution and plot based on given parameters
binomial = np.random.binomial(n = 10, p = 0.02, size = 1000)
plt.hist(binomial, range = [-2,4], density = True, alpha = .5, color = 'mediumorchid')

#poisson distribution and plot based on given parameters
poisson = poisson.rvs(mu=.2, size=1000)
plt.hist(poisson, density=True, alpha = .5, color = 'cadetblue')

#normal distribution and plot based on given parameters
normal = np.sort(norm.rvs(loc = .2, size=1000, scale =.45))
plt.plot(normal, sp.stats.norm.pdf(normal, .2, .45), color = "black")
plt.xlabel('Randomly Generated Values')
plt.ylabel('Probability')
plt.title('Binomial ($N=10, p=0.02$), Poisson ($λ=0.2$), & Normal ($μ=0.2, σ=0.45$)')
plt.legend([ 'Normal', 'Binomial', 'Poisson'])
plt.show()

# creating dataframe to hold all of the necessary information for each distribution
data1 = {
  "pdf": ["Binomial","Poisson", "Normal"],
    "Mean": [np.mean(binomial), np.mean(poisson), np.mean(normal)], 
    "Sigma": [st.variance(binomial), st.variance(poisson), st.variance(normal)],
    "Skewness": [skew(binomial), skew(poisson), skew(normal)], 
    "Kurtosis":[kurtosis(binomial),kurtosis(poisson),kurtosis(normal)]}

df1 = pd.DataFrame(data1)
print(df1) 
```

With these specified parameters, we can see very discrete values for Poisson and Binomial distributions, especially around 0. It should be noted that the binomial parameter N=10 is relatively low, as well as the Poisson mean being a low value. With the given parameters it should be expected that neither the Binomial or Poisson distributions would closely resemble that of the Normal distribution.
