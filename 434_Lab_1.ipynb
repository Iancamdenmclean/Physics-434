__(1). Calculate cumulative probabaility from mean for Z = 1, 3 and 5.__

```python
# necessary import to integrate distribution function
from scipy.stats import norm

# The cumulative density function can be utilized to calculate the probability to the left of a point, so we just use two CDF
# one is from the desired Z value to the mean, and the other is subtracting off the part from -infinity to the mean
# Z = 1
CPFmean1 = round((norm.cdf(1,0,1)- norm.cdf(0,0,1)),3)
print('The cumulative probability from the mean for Z = 1 is: ' + str(CPFmean1))

# Z = 3
CPFmean3 = round((norm.cdf(3,0,1)- norm.cdf(0,0,1)),3)
print('The cumulative probability from the mean for Z = 3 is: ' + str(CPFmean3))

# Z = 5
CPFmean5 = round((norm.cdf(5,0,1)- norm.cdf(0,0,1)),3)
print('The cumulative probability from the mean for Z = 5 is: ' + str(CPFmean5))
```

In order to calculate the cumulative probability from the mean for a given value, we can utilize the cumulative density function. This will give us the probability that a value will fall between negative infinity to our given random continuous value. In other words, to the left of the given value. So when we choose 1, the CDF gives the probability from -infinity to 1, but in order to get it from the mean to 1, we need to subtract off the CDF of -infinity to the mean (0). The parameters of the function are the chosen Z value as a continuous variable, followed by the mean, followed by the standard deviation (1). We can see that the cumulative probability from the mean increases as the Z value is increased, which intuitively makes sense. A larger area for a saple to fall in will give it a larger probability to fall into it. It is important to note that the cumulative probability from the mean for Z = 5 is essentially .5 (although rounded up) which makes sense as any value should very likely fall within 5 standard deviations of the mean, ignoring outliers. THis case just consideres the right half of a Gaussian distribution though, which is why we see .5

__(2). Calculate cumulative probability for Z = 1, 3 and 5.__

```python
# The cumulative density function can be utilized to calculate the probability to the left of a point
# we just use a single CDF in this case from -infinity to the desired Z value
# Z = 1
CPmean1 = round(norm.cdf(1,0,1),3)
print('The cumulative probability  Z = 1 is: ' + str(CPmean1))

# Z = 3
CPmean3 = round(norm.cdf(3,0,1),3)
print('The cumulative probability from the mean for Z = 3 is: ' + str(CPmean3))

# Z = 5
CPmean5 = round(norm.cdf(5,0,1),3)
print('The cumulative probability from the mean for Z = 5 is: ' + str(CPmean5))
```

The cumulative probability can be calculated simply as the cumulative density function from -infinity to a selected random value. In this case we still use Z = 1,3,5 and it should be expected for the cumulative probability values to be much larger than those taken from the mean, especially at larger Z values where esentially all of the PDF is integrated over. We can clearly see this when looking at the results as Z = 1 is already a very large probability at Z .841, whereas at 3 standard deviations from the mean, the probability is almost 1, and at 5 standard deviations the cumulative probability is basically 1 although rounded.

__(3). Calculate complementary cumulative probabaility for Z = 1, 3 and 5.__

```python
# 1 - the cumulative density function can be utilized to calculate the complementary probability
# we just use 1- the CDF in this case
# Z = 1
CCmean1 = round(1-norm.cdf(1,0,1),3)
print('The complementary cumulative probability  Z = 1 is: ' + str(CCmean1))

# Z = 3
CCmean3 = round(1-norm.cdf(3,0,1),3)
print('The complementary cumulative probability from the mean for Z = 3 is: ' + str(CCmean3))

# Z = 5
CCmean5 = round(1-norm.cdf(5,0,1),3)
print('The complementary cumulative probability from the mean for Z = 5 is: ' + str(CCmean5))
```

In order to calculate the complementary cumulative probability, we just take 1 - the values we calculated in the previous part. Conceptually, this should equal 1 when added with the cumulative probability as it should make up the entire probability density function. As we can see, the complement of the cumulative probability decreases, which makes sense with what was said earlier, as the cumulative probability increased with larger Z. Thus, most of the probability density function is contained in the cumulative density rather than the complementary cumulative density at larger Z. This is why we see a relatively small complement for Z = 1 (.159) and essentially 0 for Z =3, and 5.
