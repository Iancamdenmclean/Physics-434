__Exploratory Plots for LHC Higgs vs. QCD__

```python
# making plots for each of the keys in Higgs vs Qcd to compare general trends
# looping through each key in the dictionary to plot
for i in new_dict_higgs.keys():
    # plotting histograms for higgs and background jets
    plt.hist(new_dict_higgs[i], density = True, color = 'red', alpha = .6, bins = 40)
    plt.hist(new_dict_qcd[i], density = True, color = 'dodgerblue', alpha = .6, bins = 40)
    # adding all necessary labels
    plt.ylabel('Frequency')
    # specific key as label
    plt.xlabel(str(i))
    # updating title
    plt.title("plot of " + str(i) + " for Higgs Jets vs. QCD Jets")
    plt.legend(['Higgs Boson', 'QCD'])
    plt.show()
```

Above are histogram plots for all of the data from the dictionary. All values corresponding to each key are put into a histogram for background noise as well as Higgs boson. As we can see, the histograms are normalized and overlaid and there are some clear differences in distributions and range for these different measurements. This is helpful in determining which jet comes from a higgs boson vs background noise. When looking at each individual measurement from an unknown jet, we can look at each of the measurements corresponding to these keys to get an idea if the jet is a higgs boson or background noise.
More Specifically:

__eta:__ the plots for eta of higgs boson vs background noise are similar in the sense that the distribution is normal, however, we see that the frequency of the background noise at it's mean of eta is lower than that of the Higgs jet. The means appear to be the same.

__phi:__ The phi values are essentially indistinguishable between the Higgs jet and the background noise.

As for the Higgs Boson, it has a small, somewhat noticeable secondary peak around 0, while there is the mean at around .14 with a frequency just shy of 20. The values dip between the mean and the secondary peak.

__ee3:__ Both the Higgs boson and the background noise have sharp exponential decay form 0 to .02 for the higgs boson, ans about 0 to 0.4 for the background noise. Also, the mean of the Higgs boson has a much higher frequency, although the means of the two seem slightly similar.

__d2:__ For the Higgs boson, there is a very sharp exponential decay from 0 to 20, where the mean has a high frequency of .58 around 1 or 2. The background noise is also a decay but less drastic as the values starting from 0 and varying to 20, have more counts than that of the Higgs Boson.

__angularity:__ Plots are a similar curve sloping downward where both means are between 0 and 0.01 although the frequency at the mean for the Background noise is slightly higher. Overall the distributions are similar.

__t1:__ The plots are both similar with a mean around .85 for the Higgs boson, and only a slight shift of the plot to the right leads to the plot of the background noise. Also, the mean of the background noise has a higher frequency.

__t2:__ The distributions are very different where the Higgs boson slopes downward from 0 to 1 and a mean slightly above 0. The mean is around .75

__t3:__ For the t3 plots, they have the same range from 0 to 8, however the Higgs boson has a mean around .15 and then essentially slopes downward, whereas the backgorund noise looks like a mound with a mean around .55 but only a frequency of 2. 

__t21:__ The plots for t3 are very distinct and could likely help distinguish between Higgs Boson and background noise. The Higgs boson has a mean around .2 and slopes down after that, and then the backgorund noise increases from 0 and has a mean around .8. Then it falls off. The ranges of both of these are the same.

__t32:__ The plots of t32 show a smilar trend, however the range for the background noise is shifted slightly to the right compared to that of the Higgs boson, so this also affects the mean being shifted. However, the general plot trends are the same.

__KtDeltaR:__ For this measurement, the plots are distinct. Plot a waterfall plot of baseline (1, 2) for all the 20 time intervals.

```python
# Pick 2 sample antennas, and get data from that baseline. Note that to get the 'auto-correlations', you can do the 
# same thing, but set antenna1 and antenna2 to be the same value.
antenna1 = 1
antenna2 = 2
data_apair = uv.get_data(antenna1,antenna2)

# plotting the waterfall plot
fig, axs = plt.subplots(1,1,figsize=(12,6))
# using imshow for setting up y axis
im = plt.imshow(np.abs(data_apair),aspect='auto')
# adding heatmap
cbar = fig.colorbar(im)
# plotting frequencies for each channel
freqs = uv.freq_array[0]*1e-6 #gets frequencies associated with each channel, in MHz
# adding titles to plots
axs.set_ylabel('Time')
axs.set_xlabel('Frequency (MHz)')
# setting up x axis details
xticks = [int(i) for i in np.linspace(0,len(freqs)-1,8)]
_ = axs.set_xticks(xticks)
xticklabels = np.around(freqs[xticks],0)
_ = axs.set_xticklabels(xticklabels)
```

This is the general waterfall plot for all frequencies of each time interval. The strongest frequency is visible around 137 MHz and there are several clear distinct lines at otehr frequencies too.
