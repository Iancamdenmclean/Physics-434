1. Draw a 100k data points from a Gaussian distribution with mean μ=3 and width σ=0.1. 
In many areas of statistics a Gaussian distribution is called the 'Normal distribution'. 
So we'll dig into the "stats" package from "scipy" and use stats.norm.rvs to draw 100K points.

```python
# importing necessary packages to utilize
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
import random

# setting a random seed so that the sample array is the same each time
# using this to keep one parameter constant while working with others in the code
random.seed(30)

# pulling 100K data points from gaussian distribution of mean = 3, standard deviation = .01
# points are sorted in numerical order for plotting probability distribution function
array = np.sort(sp.stats.norm.rvs(loc = 3, size=100000, scale =.01))

#plotting probability distribution function to visualize trend
plt.plot(array, sp.stats.norm.pdf(array, 3, .1), "indigo")

#Adding title and axis labels to the plot of the trend visualization
plt.xlabel('Randomly Generated Values')
plt.ylabel('Probability Density')
plt.title('PDF of 100K randomly selected points from a Gaussian Distribution')
```

The stats package of scipy was utilized to generate 100,000 sample points from a Gaussian distribution with mean = 3 and standard deviation = .01. A seed was utilized to keep the raw data generated the same throughout the process of working with the data. This eliminates one changing parameter when focusing on changing others. The normalized distribution can be visualized above with the mean clearly at 3 as specified during the random sampling. The probability density is plotted along the y axis.

2.  Plot a histogram of these points using matplotlib. Set x-axis range 2.95 and 3.05, and number of bins 50.

```python
# making a histogram with the data and plotting it for a small range around the expectation value
plt.hist(array, bins=50, range = [2.95,3.05], color = 'coral')

# adding title and other labels to the plot
plt.xlabel('Randomly Generated Values')
plt.ylabel('Number of Occurrances')
plt.title('Histogram for PDF of 100K points from a Gaussian Distribution')
```

For the array of 100,000 randomly generated values from the Gaussian distribution, a bin width of 50 is specified and the range is constrained to +- .05 on either side of the mean. The counts of each value generated, falling within a specific bin is the y-axis. This bell curve with the centered mean clearly shows a Gaussian distribution arising from the randomly generated values.

3. Superpose a Gaussian probability density function using stats.norm.pdf() with normalization to 100K.

```python
# Previous histogram is replotted, however needs to be normalized to 100K, so we say density = True
plt.hist(array, bins=50, density=True, color = 'coral')

# Superposing normalized Gaussian probability density function onto the histogram to visualize the fit
plt.plot(array, sp.stats.norm.pdf(array, 3, .01), color = 'indigo')

# adding title and other labels to the plot
plt.xlabel('Randomly Generated Values')
plt.ylabel('Normalized Number of Occurrances')
plt.legend(['Gaussian PDF', 'Histogram'])
plt.title('Normalized Histogram vs Normalized PDF of 100K points')
```

The histogram is normalized to represent a probability density rather than specific counts, and a normalized Gaussian probability distribution function is superposed on top of the histogram. The purpose of this is to demonstrate that the randomly generated values very closely follow a normal Gaussian distribution, as the plotted probability distribution function very closely traces the shape of the histogram.

4. Draw the same histogram and PDF using log scale in y-axis.

```python
# Previous normalized histogram and PDF are replotted, 
# however needs to be log y scale so we set the scale that way
plt.yscale('log')
plt.hist(array, bins=50, density=True, color = 'coral')
plt.plot(array, sp.stats.norm.pdf(array, 3, .01), color = 'indigo')

## adding title and other labels to the plot
plt.xlabel('Randomly Generated Values')
plt.ylabel('Log of Normalized Number of Occurrances')
plt.legend(['Gaussian PDF', 'Histogram'], fontsize = 9)
plt.title('Histogram vs PDF of 100K Points with log(y) Scale')
```

The previous normalized histogram and Gaussian probability distribution function are carried over to this question, however the y-axis is set to logarothmic and the histogram clearly still follows the trend of the Gaussian PDF. Making a logarithmic y-axis allows us to better compare each bin's behavior at different points.
